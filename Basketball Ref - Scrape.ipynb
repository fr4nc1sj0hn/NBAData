```python
import pandas as pd
url = "https://www.basketball-reference.com/play-index/pgl_finder.cgi?request=1&player_id=&match=game&year_min=2001&year_max=2018&age_min=0&age_max=99&team_id=&opp_id=&season_start=1&season_end=-1&is_playoffs=N&draft_year=&round_id=&game_num_type=&game_num_min=&game_num_max=&game_month=&game_day=&game_location=&game_result=&is_starter=&is_active=Y&is_hof=&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&pos_is_c=Y&pos_is_cf=Y&c1stat=&c1comp=&c1val=&c1val_orig=&c2stat=&c2comp=&c2val=&c2val_orig=&c3stat=&c3comp=&c3val=&c3val_orig=&c4stat=&c4comp=&c4val=&c4val_orig=&is_dbl_dbl=&is_trp_dbl=&order_by=date_game&order_by_asc=Y&offset="
start = 0
end = 10000
step = 100
for offset in range(start,end,step):
    try:
        link = url + str(offset)
        lst = pd.read_html(link)
        df = lst[0]
        stat = df[df["Rk"] != "Rk"]
        print(str(offset))
        stat.to_csv("C:/My/pyprojects/NBA/Active01.csv",mode='a',index=False)
    except ValueError:
        print("Error")
```
